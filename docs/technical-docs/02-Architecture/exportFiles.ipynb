```python
import os
src_dir = "./Entities"

files = []
for root, dirs, filenames in os.walk(src_dir):
    for f in filenames:
        files.append(f)
```

```python
files
```

```python
import pandas as pd
```

```python
files = pd.DataFrame(files, columns=["filename"])
files.to_csv("files.txt", index=False, header=False)
```
